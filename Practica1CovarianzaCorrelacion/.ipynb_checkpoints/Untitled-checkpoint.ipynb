## Importacion de librerias

```python
#Librería para manipulación y estructura de datos
import pandas as pd
# Librería de visualización de datos
import matplotlib.pyplot as plt
# Librería basada en matplotlib que facilita la creación de gráficos estadsticos
import seaborn as sns
```

```python
data=pd.read_csv('MentalHealthSurvey.csv')
data
```

```python
data.info()
```

## Seleccion de variables

-gender
-age
-degree major
-academic_workload
-social_relationships
-depression
-anxiety
-isolation

```python
selected_columns = ['gender', 'age', 'academic_workload ', 'financial_concerns', 'social_relationships', 'depression', 'anxiety', 'isolation']
data = data[selected_columns]
data
```

## Distribucion de ansiedad y depresion

```python
plt.figure()
sns.histplot(data['anxiety'], kde=True)
plt.title("Distribucion de anxiety")
plt.xlabel("Nivel de ansiedad")
plt.ylabel("frecuencia")
plt.show()
```

```python
plt.figure()
sns.histplot(data['depression'], kde=True)
plt.title("Distribucion de depression")
plt.xlabel("Nivel de depresion")
plt.ylabel("frecuencia")
plt.show()
```

## Medidas de tendencia central

```python
data.describe()
```

```python
g = sns.catplot(
    data=data, kind="bar", x="age",y="anxiety", hue="gender")
g.despine(left=True)
g.set_axis_labels("", "anxiety level")
g.legend.set_title("")
```

```python
plt.figure()
sns.boxplot(x='anxiety', y="academic_workload ", data=data)
plt.title("anxiety vs academic_workload")
plt.xlabel("anxiety")
plt.ylabel("Sale Price")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='anxiety', y="financial_concerns", data=data)
plt.title("anxiety vs financial_concerns")
plt.xlabel("anxiety")
plt.ylabel("financial_concerns")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='isolation', y="social_relationships", data=data)
plt.title("isolation vs social_relationships")
plt.xlabel("isolation")
plt.ylabel("social_relationships")
plt.show()
```

```python
selected_columns = ['age', 'academic_workload ', 'financial_concerns', 'social_relationships', 'depression', 'anxiety', 'isolation']
data2 = data[selected_columns]
data2
```

## Matriz de covarianzas y correlacion

```python
cov_matrix = data2.cov()
corr_matrix = data2.corr()
```

```python
cov_matrix
```

```python
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(data2, kind='reg',plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.5}})
plt.title("Pairplot")
plt.show()
```
